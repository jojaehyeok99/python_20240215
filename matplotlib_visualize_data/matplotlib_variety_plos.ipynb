### 바 차트 (막대 그래프)
x 축의 값이 분류가 가능한 카테고리 형태일 때 `bar()`, `barh()` 메서드로 막대그래프를 표시할 수 있음  

주의! bar 메서드는 x축의 값이 첫번째 인수로 옴

```python
y = [2, 3, 1]
x = [1, 2, 3]
xlabel = ['가', '나', '다']

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.show()
```

`xerr` 인수나 `yerr` 인수를 지정하면 에러 바(error bar)를 추가할 수 있음  

`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음 (0 : 완전 투명, 1 : 완전 불투명)

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title("Bar h")
plt.barh(y_position, y_performance, xerr=error, alpha=0.4)
plt.yticks(y_position, people)
plt.xlabel('x 라벨')
plt.grid(True)
plt.show()
```

### 스템 플롯
이산활륙이나 자기 상관관계를 표현하고자 할 때 사용하는 차트  

`stem()` 메서드로 표시할 수 있음

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x), '-.')
plt.grid(True)
plt.show()
```

### 파이 차트
각 분류가 전체에서 얼마만큼 차지하고 있는지 파악하려 할 때 사용하는 차트  

`pie()` 메서드로 표현할 수 있음

주의! 차트가 원형태를 유지할 수 있게 하려면 `axis('equal')`을 지정해야 함

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] # RGB는 안됨
explode = (0, 0.1, 0, 0) # 부분 중 하나를 강조하기 위해 튀어나오게 만듦, 시계 반대방향 순서
plt.title('파이 차트')
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) 
        # %1.1f%%는 소수점 이하 한 자리까지의 퍼센트 값을 표시하도록 지정한 것
plt.axis('equal')
plt.show()
```
